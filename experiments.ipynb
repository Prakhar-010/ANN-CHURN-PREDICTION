```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
import pickle
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
df.tail()
```

```python
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df[['Gender']])
df
```

```python
label_encoder_gender.classes_
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Step 1: Fit and transform
onehot_encoder_geography = OneHotEncoder()
encoded = onehot_encoder_geography.fit_transform(df[['Geography']]).toarray()

# Step 2: Create dataframe with proper column names
encoded_df = pd.DataFrame(
    encoded, 
    columns=onehot_encoder_geography.get_feature_names_out(['Geography'])
)

# Step 3: Drop original column and join encoded columns
df = df.drop('Geography', axis=1)
df = pd.concat([df, encoded_df], axis=1)

print(df.head())
```

```python
df.head()
```

```python
with open('label_encoder_gender.pkl','wb')as file:
    pickle.dump(label_encoder_gender,file)
with open('onehot_encoder_gender.pkl','wb')as file:
    pickle.dump(onehot_encoder_geography,file)
```

```python
##divide the dataset
df.head()
```

```python
x = df.drop('Exited',axis=1)
y = df['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train.shape[1]
```

```python
with open('scalar.pkl','wb')as file:
    pickle.dump(scalar,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
##compile the parameter
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])##for multi classification use spare classentropy
```

```python
##set up the tensorboard
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir="logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=tensorflow_callback = tf.keras.callbacks.TensorBoard(
    log_dir='logs',
    histogram_freq=1
)
```

```python
##set up early stopping
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
!pip install tensorboard
```

```python
##train the model
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,callbacks=[tensorflow_callback,early_stopping_callback]
                  )
```

```python
model.save('model.h5')
```
